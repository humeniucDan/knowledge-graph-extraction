```python
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline
```

```python
MODEL_DIR = "../train/out/pos+ner_distilbert-base-uncased/checkpoint-6250"
```

```python
tokenizer = AutoTokenizer.from_pretrained(MODEL_DIR)
model = AutoModelForTokenClassification.from_pretrained(MODEL_DIR)
model.eval()
ner = pipeline("token-classification", model=model, tokenizer=tokenizer, aggregation_strategy="none")
```

```python
text = "Rocky, our dog, has been chasing that his tail for hours."
results = ner(text)
for r in results:
    print(r)
    # print(r['word'], r['entity_group'], r['score'])
```
